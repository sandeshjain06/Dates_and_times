```python
date['date_year']=date['date'].dt.year
date.sample(5)
```

```python
date['date_month']=date['date'].dt.month
date.sample(5)
```

```python
date['date_month_name']=date['date'].dt.month_name()
date.sample(5)
```

```python
date['date_day'] =date['date'].dt.day
date.sample(5)
```

```python
date['date_day_ofweek'] =date['date'].dt.dayofweek
date.sample(5)
```

```python
date['date_week_name'] =date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# whether it is weekend or not

date['date_is_weekend']=np.where(date['date_week_name'].isin(['Sunday','Saturday']),1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```
